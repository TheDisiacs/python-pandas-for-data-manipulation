# `.between()`

### Import pandas

```python
# import pandas
import pandas as pd
from pathlib import Path
path = Path('C:/Users/alouden01/Documents/python-pandas-for-data-manipulation/starter_files/skill_10_filtering_duplicates/employee.csv')
```

### Employee data


```python
# read csv and parse the date column
emp = pd.read_csv(path, parse_dates=['start_date'])
```

### Head

```python
# called the head method to see what were working with
emp.head()
```

### Manually using the `&` operator

```python
mask_1 = emp['salary'] > 300000
mask_2 = emp['salary'] < 600000
emp[mask_1 & mask_2].sort_values(by=['salary'])
```

### The `between()` Method

```python
# inclusive
sal_mask = emp['salary'].between(300000, 600000)
```

### Extract by Passing `mask` to `df[]`

```python
# return masked results
emp[sal_mask]
```

```python
# return masked results: sorted by salary
emp[sal_mask].sort_values(by='salary')
```

### Between dates

```python
date_mask = emp['start_date'].between('1/1/2020','12/31/2020')
```

```python
# return masked results
emp[date_mask]
```

```python
# sort by date
emp[date_mask].sort_values(by='start_date')
```
