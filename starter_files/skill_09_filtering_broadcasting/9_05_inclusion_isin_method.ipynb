### The `isin()` Method

```python
data_mask = emp['team'].isin(['data analytics'])
management_mask = emp['team'].isin(['management'])
emp[data_mask | management_mask]
```

### Extract by Passing `mask` to `df[]`

```python
# female managers
female_mask = emp['gender'].isin(['Female'])
emp[female_mask & management_mask]
```

```python
olly_mask = emp['first_name'].isin(['Olly'])
remote_mask = emp['remote'].isin([True])
```

```python
emp[female_mask][remote_mask][management_mask]
```
