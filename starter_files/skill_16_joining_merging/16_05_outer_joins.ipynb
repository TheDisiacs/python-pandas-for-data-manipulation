```python

from pathlib import Path
root = Path('C:/Users/alouden01/Documents/python-pandas-for-data-manipulation/starter_files/skill_16_joining_merging/cafe_customers.csv').parent
customers = pd.read_csv(root / 'cafe_customers.csv')
items = pd.read_csv(root / 'cafe_items.csv')
week1= pd.read_csv(root / 'cafe_week_1_sales.csv')
week2= pd.read_csv(root / 'cafe_week_2_sales.csv')
```

```python
# perform an outer join on week_1 and week_2
# week1.merge(week2, how='outer', on='cust_id', suffixes=['_wk1','_wk2']).dropna(how='any')
week1.merge(week2, how='outer', on='cust_id', suffixes=('_wk1','_wk2')).head(20)
```

```python
# cust_id 3 is an inner join result while cust_id 0 only purchased in week_1
# add an indicator
outer = week1.merge(
    week2,
    how='outer',
    on='cust_id',
    suffixes=('_w1','_w2'),
    indicator=True
)
outer
```

```python
full_outer_mask = outer['_merge'].isin(['left_only','right_only'])
outer[full_outer_mask]
```
