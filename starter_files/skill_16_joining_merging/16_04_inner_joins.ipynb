### Perform an inner join on week_1 and week_2 sales

```python
# only returns customers who purchased in week_1 AND week_2
# We are using df.merge
from numpy import inner


both_weeks = week1.merge(week2,'inner','cust_id')
both_weeks
```

```python
both_weeks = week1.merge(week2, how='inner', on='cust_id', suffixes=('_w1','_w2'))
both_weeks
```

```python
# note the duplicate cust_ids
# merge on multiple things cust_id + item_id
multi = week1.merge(week2, how='inner', on=['cust_id', 'item_id'])
multi
```
